```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
diabetes = pd.read_csv("diabetes_.csv")
print(diabetes.head())
```

```python
diabetes.info()
```

```python
diabetes.describe()
```

```python
diabetes=diabetes.drop_duplicates()
```

```python
print(diabetes.shape)
```

```python
diabetes.isnull().sum()
```

```python
print('No. of zero values in Glucose is',diabetes[diabetes['Glucose']==0].shape[0])
print('No. of zero values in BloodPressure is',diabetes[diabetes['BloodPressure']==0].shape[0])
print('No. of zero values in SkinThickness is',diabetes[diabetes['SkinThickness']==0].shape[0])
print('No. of zero values in Insulin is',diabetes[diabetes['Insulin']==0].shape[0])
print('No. of zero values in BMI is',diabetes[diabetes['BMI']==0].shape[0])
```

```python
# PREGUNTA PROBLEMA ?¿
diabetes['Glucose']=diabetes['Glucose'].replace(0,diabetes['Glucose'].mean())
diabetes['BloodPressure']=diabetes['BloodPressure'].replace(0,diabetes['BloodPressure'].mean())
diabetes['SkinThickness']=diabetes['SkinThickness'].replace(0,diabetes['SkinThickness'].mean())
diabetes['Insulin']=diabetes['Insulin'].replace(0,diabetes['Insulin'].mean())
diabetes['BMI']=diabetes['BMI'].replace(0,diabetes['BMI'].mean())
```

```python
diabetes.describe()
```

```python
diabetes["Outcome"].value_counts().to_frame()
```

```python
diabetes["Outcome"].value_counts(normalize=True).to_frame()
```

```python
# Histograma de cada característica
diabetes.hist(figsize = (10,10),bins=10)
plt.show()
```

```python
# Obtener la correlación de cada característica en el dataset
cormat = diabetes.corr()
top_corr_features = cormat.index
plt.figure(figsize=(20,20))
#Obtener el mapa de calor
g = sns.heatmap(diabetes[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
# Definir las características
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# Crear los histogramas
for feature in features:
    plt.figure(figsize=(10, 5))
    diabetes[diabetes['Outcome'] == 0][feature].plot(kind='hist', alpha=0.5, color='blue', bins=30, label='Outcome 0')
    diabetes[diabetes['Outcome'] == 1][feature].plot(kind='hist', alpha=0.5, color='red', bins=30, label='Outcome 1')
    plt.title(f'Histograma de {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frecuencia')
    plt.legend()
    plt.show()
```

```python
import plotly.graph_objs as go
import plotly.express as px

# Crear la figura
fig = go.Figure()

# Agregar trazos para cada clase
fig.add_trace(go.Scatter3d(
    x=diabetes[diabetes['Outcome'] == 0]['Glucose'],
    y=diabetes[diabetes['Outcome'] == 0]['BMI'],
    z=diabetes[diabetes['Outcome'] == 0]['Pregnancies'],
    mode='markers',
    marker=dict(size=5, color='blue'),
    name='Outcome 0'
))

fig.add_trace(go.Scatter3d(
    x=diabetes[diabetes['Outcome'] == 1]['Glucose'],
    y=diabetes[diabetes['Outcome'] == 1]['BMI'],
    z=diabetes[diabetes['Outcome'] == 1]['Pregnancies'],
    mode='markers',
    marker=dict(size=5, color='red'),
    name='Outcome 1'
))

# Actualizar diseño del gráfico
fig.update_layout(
    title='3D Scatter Plot of Glucose, BMI, and Pregnancies by Outcome',
    scene=dict(
        xaxis_title='Glucose',
        yaxis_title='BMI',
        zaxis_title='Pregnancies'
    )
)

# Mostrar el gráfico
fig.show()
```
